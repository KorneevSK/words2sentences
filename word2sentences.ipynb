# Генерация текста на основе вероятности слова

```python
dict_words = {}
```

```python
for i in range(len(df)):
    dict_words.update({df.Word[i]: df.Probability[i]})
```

```python
dict_words
```

```python
def choose(d_w):
    total = 0
    for k,v in d_w.items():
        total += v
    randomPoint = np.random.random() * total
    for k,v in d_w.items():
        if randomPoint < v:
            return k
        else:
            randomPoint -= v
    return len(d_w) - 1
```

```python
n = int(input('Введите длину предложения: '))
for i in range(n):
    print(choose(dict_words),end=' ')
```

# Генерация цепью Маркова

```python
word_list = list(df['Word'])
```

```python
from collections import defaultdict
def markov_chain(text):
    m_dict = defaultdict(list)
    for current_word, next_word in zip(text[0:-1], text[1:]):
        m_dict[current_word].append(next_word)
    m_dict = dict(m_dict)
    return m_dict
```

```python
word_dict = markov_chain(word_list)
word_dict
```

```python
def generate_sentence(chain, count=12):
    word1 = np.random.choice(list(chain.keys()))
    sentence = word1.capitalize()

    for i in range(count-1):
        word2 = np.random.choice(chain[word1])
        word1 = word2
        sentence += ' ' + word2

    sentence += '.'
    return(sentence)
```

```python
generate_sentence(word_dict)
```
